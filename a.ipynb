```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv("fm.csv")

# Filter rows where the "MH_id" column equals 171377

df.head()
```

```python
ykb = df[df["PH_City"] == 171377]

print("Unique MH_id values:", df["MH_id"].unique())

print("Number of rows in ykb DataFrame:", len(ykb))

# Print the first few rows of the filtered DataFrame
print("First few rows of ykb DataFrame:")
print(ykb.head())

# Write the filtered DataFrame to a CSV file
ykb.to_csv('ykb.csv', index=False)
```
